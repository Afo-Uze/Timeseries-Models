```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRFRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
path = '/content/drive/MyDrive/Salesreport.csv'
Sales = pd.read_csv(path)
```

```python
Sales.head(10)
```

Checking for nulls values in the dataset


```python
Sales.info()
```

Dropping store and item columns

```python
Sales = Sales.drop(['store','item'], axis=1)
```

```python
Sales.info()
```

Converting date from object datatype to datatime datatype

```python
Sales['date']=pd.to_datetime(Sales['date'])
```

```python
Sales.info()
```

converting date to a month period, and then sum the number of items in each month

```python
Sales['date'] = Sales['date'].dt.to_period("M")
monthly_sales = Sales.groupby('date').sum().reset_index()
```

converting the resulting date to timestamp datetype

```python
monthly_sales['date'] = monthly_sales['date'].dt.to_timestamp()
```

```python
monthly_sales.head(10)
```

Visualization of the monthly_sales

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'],monthly_sales['sales'])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Monthly Customer Sales")
plt.show()
```

Call the difference bewteen sales columns to make the sales date stationery

```python
monthly_sales['sales_diff'] = monthly_sales['sales'].diff()
monthly_sales = monthly_sales.dropna()
monthly_sales.head(10)
```

Viulazation of sales and sales_diff

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Monthly  Customer Sales Difference")
plt.show()
```

Dropping off Sales and date


```python
supervised_data = monthly_sales.drop(['date','sales'], axis=1)
```

Preparing the supervised data

```python
for i in range(1,13):
  col_name = 'month_' + str(i)
  supervised_data[col_name]= supervised_data['sales_diff'].shift(i)
supervised_data = supervised_data.dropna().reset_index(drop=True)
supervised_data.head(10)
```

Slipt data into train and test

```python
train_data = supervised_data[:-12]
test_data = supervised_data[-12:]
print("Train Data Shape: ", train_data.shape )
print("Test Data Shape: ", test_data.shape )
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))
scaler.fit(train_data)
train_data = scaler.transform(train_data)
test_data = scaler.transform(test_data)
```

```python
X_train, Y_train = train_data[:,1:], train_data[:,0:1]
X_test, Y_test = test_data[:,1:], test_data[:,0:1]
Y_train = Y_train.ravel()
Y_test = Y_test.ravel()
print("x_train Shape", X_train.shape)
print("y_train Shape", Y_train.shape)
print("x_test Shape", X_test.shape)
print("y_test Shape", Y_test.shape)
```

Create prediction dataframe to merge the predicted sales of all trained alogrimths

```python
Sales = monthly_sales['date'][-12:].reset_index(drop=True)
predict_df = pd.DataFrame(Sales)
```

```python
Act_sales = monthly_sales['sales'][-13:].to_list()
print(Act_sales)
```

To create the linear regression model and predicted output

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
lin_pre = lin_model.predict(X_test)

```

```python
lin_pre = lin_pre.reshape(-1,1)
#this is a set matrix,and it contains the input features of the data,and also the predicted output
lin_pre_test_set = np.concatenate([lin_pre, X_test], axis=1)
lin_pre_test_set = scaler.inverse_transform(lin_pre_test_set)
```

```python
result_list =[]
for  index in range(0, len(lin_pre_test_set)):
  result_list.append(lin_pre_test_set[index][0]+ Act_sales[index])
lin_pre_series = pd.Series(result_list, name="Linear Prediction")
predict_df = predict_df.merge(lin_pre_series, left_index = True, right_index = True)
```

```python
print(predict_df)
```

```python
lin_mse = np.sqrt(mean_squared_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:]))
lin_mae = mean_absolute_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
lin_r2 = r2_score = (predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
print("Linear Regression MSE: ", lin_mse)
print("Linear Regression MAE: ", lin_mae)
print("Linear Regression R2: ", lin_r2)
```

visualization of the prediction against the actual sales

```python
plt.figure(figsize=(15,5))
# Actual sales
plt.plot(monthly_sales['date'], monthly_sales['sales'])
# predicted sales
plt.plot(predict_df['date'], predict_df['Linear Prediction'])
plt.title("customer sales Forecast Using Linear Regression Model")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend(['Actual Sales', 'Predicated sales'])
plt.show
```
